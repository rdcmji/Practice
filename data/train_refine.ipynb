```python
import re

diffre = re.compile('[0-9]*[a-zA-z]+[0-9]+[a-zA-Z]*\s[0-9]*[a-zA-z]*[0-9]+[a-zA-Z]*')
num_data =3000
diffnum=[]
diffline=[]

with open('train.66846.msg') as reader:
    trainmsg = reader.readlines()
    for i, j in enumerate(trainmsg):
        temp = re.search(diffre, j)
        if temp:
            #print (j)
            #print(temp)
            diffnum.append(i)
            diffline.append(j)
with open ('train.66846.diff') as reader:
    traindiff = reader.readlines()
print (diffnum)
print(diffline)
print(len(diffline))
```

```python
final_msg =[trainmsg[i] for i in range(len(trainmsg)) if i not in diffnum]
final_diff = [traindiff[i] for i in range(len(trainmsg)) if i not in diffnum]
print (len(final_msg))
```

```python
print(len(final_msg))
```

```python
#msg  
for i in range(len(final_msg)):
    final_msg[i] = final_msg[i].lower()
```

```python
#ignore msg index list 
nig=[]
for i in range(len(final_msg)):
    if final_msg[i].find("ignore") == -1:
        nig.append(i)
print(len(nig))
```

```python
# diff msg 
f_train_diff = [final_diff[i] for i in nig]
f_train_msg = [final_msg[i] for i in nig]
```

```python
print(len(f_train_diff))
print(len(f_train_msg))
```

```python
out_diff_name = "train.%d.diff"%(len(f_train_diff))
out_msg_name = "train.%d.msg"%(len(f_train_msg))

with open (out_diff_name, 'w') as writer:
    writer.writelines(f_train_diff)
    
with open (out_msg_name, 'w') as writer:
    writer.writelines(f_train_msg)
```
